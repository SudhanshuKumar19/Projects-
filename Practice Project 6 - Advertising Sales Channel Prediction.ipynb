```python
import numpy as np
ad_proj.skew()
for col in ad_proj.skew().index:
    if col in ad_proj.columns:
        if ad_proj.skew().loc[col]>0.5:
            ad_proj[col] = np.sqrt(ad_proj[col])
        if ad_proj.skew().loc[col]<-0.5:
            ad_proj[col] = np.cbrt(ad_proj[col])
            
ad_proj.skew()
```

# Checking Multicollinearity ---

```python
x = ad_proj.iloc[:,:-1]
y = ad_proj.iloc[:,-1]
print(x.shape,y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
stand_scale = StandardScaler()
x = stand_scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error

train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = 43, test_size = 0.20)
```

```python
def fun(f):
    f.fit(train_x,train_y)
    pred = f.predict(test_x)
    print("Mean Squared Error =",mean_squared_error(test_y,pred))
    print("r2 score =",r2_score(test_y,pred))
```

# Cross Validation Score ---

```python
from sklearn.model_selection import cross_val_score,KFold

def cvs(m):
    score = cross_val_score(m, x, y, cv = 5)
    print("Score =", score)
    print("Mean Score =", score.mean())
```

```python
lin_reg = LinearRegression()
fun(lin_reg)
```

```python
cvs(lin_reg)
```
